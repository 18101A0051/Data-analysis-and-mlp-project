```python
#to use 1.5IQR rule we have to arrange the data in ascending order
```

```python
data_n=data1.sort_values(by='Year')
```

```python
data_n.head()
```

```python
q1=data_n.quantile(0.25)#first quartile
q3=data_n.quantile(0.75)#third quartile

```

```python
IQR=q3-q1#Inner quartile range
```

```python
IQR
```

```python
#Implementing 1.5IQR rule and therby removing outliers
data_n=data_n[-((data_n<(q1-1.5*IQR)) | (data_n>(q3+ 1.5*IQR))).any(axis=1)]
```

# Machine learning 

## a) Linear Regression

```python
#dropping any left nan values
data_n.dropna(inplace=True)
```

```python
X=data_n.drop(["Price"],axis=1)
```

```python
y=data_n['Price']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lm=LinearRegression()#creating an object
```

```python
lm.fit(X_train,y_train)#training the data
```

```python
pred=lm.predict(X_test)#predicting
```

```python
# The below scatterplot shows that our prediction algorithms were true to large extent
```

```python
plt.scatter(y_test,pred)
```

```python
from sklearn import metrics
print("mean abs error",metrics.mean_absolute_error(y_test,pred))
print("mean square error",metrics.mean_squared_error(y_test,pred))
print("root mean square error",np.sqrt(metrics.mean_squared_error(y_test,pred)))

```
